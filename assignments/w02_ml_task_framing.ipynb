## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary success metric is Precision@K (e.g., Precision@50). This metric is  more valuable than Accuracy, which could be artificially inflated on a large dataset (where most pages are fine) and fail to highlight actionable opportunities. Precision@K tells us exactly what fraction of the top K pages flagged by the model genuinely require a content refresh, providing highly informative and practical value for a team with limited review capacity.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Simple "if-then" rules have limited capabilities because they rely on rigid thresholds and hard boundaries. A fixed heuristic is often brittle and treats a page just barely missing a cutoff identically to an extreme outlier. Machine Learning, on the other hand, excels here because it can learn and capture complex, non-linear relationships across multiple features simultaneously. This allows an ML model to dynamically weigh various signals—such as search volume, content age, and click-through rate—separating true structural decline from low-volume noise much more effectively than hardcoded rules.
